```python
  import pandas as pd

  df = pd.read_csv('../data/greenhouse_gas.csv')
  df.head()
```

```python
  df.info()
  df.isnull().sum()
```

```python
  # Fill numeric columns with mean, categorical with mode
  for col in df.columns:
      if df[col].dtype == 'object':
          df[col].fillna(df[col].mode()[0], inplace=True)
      else:
          df[col].fillna(df[col].mean(), inplace=True)
```

```python
# Basic statistics
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
df['Supply Chain Emission Factors without Margins'].hist()
plt.title('Emission Factor Distribution')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Boxplot for outliers
sns.boxplot(x=df['Supply Chain Emission Factors without Margins'])
plt.title('Boxplot of Emission Factor')
plt.show()
```

```python
# Countplot for categorical features
for col in ['Commodity Name', 'Substance', 'Unit']:
    plt.figure(figsize=(10,4))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.show()
```
